```python
dados = pd.read_csv('data/Spotify Most Streamed Songs.csv')
dados['in_shazam_charts'] = pd.to_numeric(dados['in_shazam_charts'], errors='coerce')
dados['in_shazam_charts'].fillna(0, inplace=True)
dados['in_shazam_charts'] = dados['in_shazam_charts'].astype(int)
dados.drop(columns=['key'], inplace=True)
dados['release_date'] = pd.to_datetime(
    dados['released_day'].astype(str) + '-' +
    dados['released_month'].astype(str) + '-' +
    dados['released_year'].astype(str),
    dayfirst=True,
    errors='coerce'
)
dados.to_csv('data/Spotify Most Streamed Songs_EDA.csv', index=False)
```

```python
import pandas as pd
import numpy as np
dados = pd.read_csv('data/Spotify Most Streamed Songs_EDA.csv')

x = dados['streams']
y = dados.select_dtypes(include=['int64']).columns
y = y[y != 'streams']

def calculate_covariances(x, y):
    covariances = {}
    for col in y:
        cov_value = x.cov(dados[col])
        covariances[col] = cov_value
    return covariances

covariances = calculate_covariances(x, y)
for col, cov in covariances.items():
    print(f'Covariância entre streams e {col}: {cov:.4f}')
```
