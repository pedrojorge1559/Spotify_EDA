```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dados = pd.read_csv('data/Spotify Most Streamed Songs_EDA.csv')
#setar variavel y (resposta)
resposta = 'streams'
#contar quantas variaveis se encaixam em int64 e não sejam a variavel y
int64_vars = dados.select_dtypes(include=['int64']).columns
int64_vars = int64_vars[int64_vars != resposta]
#função para plotar o grafico de dispersão para cada variavel int64 em x e a variavel resposta em y
def plot_scatter(variavel):
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=variavel, y=resposta, data=dados)
    plt.title(f'Relação entre {variavel} e {resposta}')
    plt.xlabel(variavel)
    plt.ylabel(resposta)
    plt.savefig(f'graphs/scatter_plot/{variavel} vs {resposta}.png')
    plt.show()
#chamar a função para printar todos os graficos de dispersão
for variavel in int64_vars:
    plot_scatter(variavel)
```
